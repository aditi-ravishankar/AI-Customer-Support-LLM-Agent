```python
import os
import re

def replace_indigo_with_skywings(directory_path):
    # Pattern to match "Indigo" regardless of case and any characters before or after it
    pattern = re.compile(r'indigo', re.IGNORECASE)
    
    # Check if directory exists
    if not os.path.exists(directory_path):
        print(f"Directory {directory_path} does not exist.")
        return
    
    # Get list of text files in the directory
    files = [f for f in os.listdir(directory_path) if f.endswith('.txt')]
    
    if not files:
        print(f"No text files found in {directory_path}.")
        return
    
    files_processed = 0
    replacements_made = 0
    
    # Process each file
    for file_name in files:
        file_path = os.path.join(directory_path, file_name)
        
        try:
            # Read file content
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
            
            # Count replacements and replace "Indigo" with "SkyWings"
            new_content, count = pattern.subn('SkyWings', content)
            replacements_made += count
            
            # Write modified content back to file
            if count > 0:
                with open(file_path, 'w', encoding='utf-8') as file:
                    file.write(new_content)
                print(f"Processed {file_name}: {count} replacements")
            else:
                print(f"Processed {file_name}: No replacements needed")
                
            files_processed += 1
                
        except Exception as e:
            print(f"Error processing {file_name}: {str(e)}")
    
    print(f"\nSummary:")
    print(f"Files processed: {files_processed}")
    print(f"Total replacements made: {replacements_made}")

if __name__ == "__main__":
    # Path to your directory containing the text files
    directory_path = r"D:\Ranjith\IISc Computational Data Science\Capstone\Airline Policy Documents"
    
    replace_indigo_with_skywings(directory_path)
```
